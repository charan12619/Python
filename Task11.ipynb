```python
import ipywidgets as widgets
from IPython.display import display

# Create label (like tkinter.Label)
label = widgets.Label(value="Hello, World!")

# Create button (like tkinter.Button)
button = widgets.Button(description="Change Font")

# Define button click event
def change_font(event):
    label.value = "Hello, World! (Font changed to bold & bigger)"

# Link the button to the function
button.on_click(change_font)

# Display the label and button
display(label, button)
```
